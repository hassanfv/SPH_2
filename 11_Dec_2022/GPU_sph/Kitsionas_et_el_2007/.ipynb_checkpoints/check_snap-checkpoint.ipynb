```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
unitTime_in_Myrs = 0.5048 # Myrs
UnitDensity_in_cgs = 5.9e-20
```

```python
jj = 60
file = h5py.File('./output_Gad_EOS_Arreaga/snap_' + str(jj).zfill(3) +'.hdf5', 'r')
```

```python
tStep = 0.005 # From Gadget parameter file.
currentTime = jj * tStep * unitTime_in_Myrs
print(f'currentTime = {round(currentTime, 3)} Myrs.')
```

### Find the number of particles inside a sphere of radius `h`

```python
dx = r[j, 0] - r[:, 0]
dy = r[j, 1] - r[:, 1]
dz = r[j, 2] - r[:, 2]

dd = (dx*dx + dy*dy + dz*dz)**0.5

print(sum(dd <= 2*h))
```

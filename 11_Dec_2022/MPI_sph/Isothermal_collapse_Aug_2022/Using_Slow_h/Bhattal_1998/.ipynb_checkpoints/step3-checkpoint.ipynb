## Intercloud particle

```python
dx = dy = dz = 0.12

xgrid = np.arange(-3, 3, dx)
ygrid = np.arange(-2, 2, dy)
zgrid = np.arange(-2, 2, dz)

res = []

for i in range(len(xgrid)):
    for j in range(len(ygrid)):
        for k in range(len(zgrid)):
            
            res.append([xgrid[i], ygrid[j], zgrid[k]])

res = np.array(res)
print(res.shape)
```

```python
# This will be used by the find_best_dx code.
mtmp = np.zeros(res.shape[0]) + m[0]
dictx = {'r': res, 'm': mtmp}
with open('ICM.pkl', 'wb') as f:
    pickle.dump(dictx, f)
```

## Note:<br>
We calculate the distance of all particles in the grid from the center of the first clump at (-1, 0, 0), and then remove all particles located at distance <= 1.0. Then we do the same for the second particle with its center located at (1, 0, 0).

```python
xcen1 = [-1, 0, 0]
dist1 = ((res[:, 0]-xcen1[0])**2 + (res[:, 1]-xcen1[1])**2 + (res[:, 2]-xcen1[2])**2)**0.5
nx1 = np.where(dist1 > 1.0)[0]

res = res[nx1, :]
print(res.shape)
```

```python
xcen2 = [1, 0, 0]
dist2 = ((res[:, 0]-xcen2[0])**2 + (res[:, 1]-xcen2[1])**2 + (res[:, 2]-xcen2[2])**2)**0.5
nx2 = np.where(dist2 > 1.0)[0]

res = res[nx2, :]
print(res.shape)
```

```python
plt.figure(figsize = (9, 7))
plt.scatter(res[:, 0], res[:, 1], s = 0.2);
```

```python
x = res[:, 0]
y = res[:, 1]
z = res[:, 2]
plt.figure(figsize = (6, 5)).add_subplot(111, projection='3d').scatter(x, y, z, s = .05);
```

## Placing the first clump into the grid

```python
# First puting the center at (-1, 0, 0)

rclump1 = r.copy()
rclump1[:, 0] = rclump1[:, 0] - 1.0 # Shifting along the x-axis

plt.figure(figsize = (6, 5))
plt.scatter(rclump1[:, 0], rclump1[:, 1], s = 0.2, color = 'k');
```

```python
# Appending rclump1 with res
res1 = np.concatenate((res, rclump1), axis = 0)
```

```python
plt.figure(figsize = (11, 7))
plt.scatter(res1[:, 0], res1[:, 1], s = 0.2, color = 'k')
plt.xlim(-3.3, 3.3)
plt.ylim(-2.3, 2.3);
```

# Appending the second clump

```python
# First puting the center at (-1, 0, 0)

rclump2 = r.copy()
rclump2[:, 0] = rclump2[:, 0] + 1.0 # Shifting along the x-axis

plt.figure(figsize = (6, 5))
plt.scatter(rclump2[:, 0], rclump2[:, 1], s = 0.2, color = 'k');
```

```python
# Appending rclump1 with res
resx = np.concatenate((res1, rclump2), axis = 0)
```

```python
plt.figure(figsize = (11, 7))
plt.scatter(resx[:, 0], resx[:, 1], s = 0.2, color = 'k')
plt.xlim(-3.3, 3.3)
plt.ylim(-2.3, 2.3);
```

```python
Npart = 2 * r.shape[0]

Ntot = resx.shape[0]

print('Number of total particles in 2 clumps = ', Npart)
print('Total particles = ', Ntot)
```

# Defining the rigid edges

```python
min(xgrid), max(xgrid)
```

```python
nxB1 = np.where(resx[:, 0] == min(xgrid))[0]
nxB2 = np.where(resx[:, 0] == max(xgrid))[0]

nyB1 = np.where(resx[:, 1] == min(ygrid))[0]
nyB2 = np.where(resx[:, 1] == max(ygrid))[0]

nzB1 = np.where(resx[:, 2] == min(zgrid))[0]
nzB2 = np.where(resx[:, 2] == max(zgrid))[0]

# Note that there are naturally some common particles in different nxB.
len(nxB1), len(nxB2), len(nyB1), len(nyB2), len(nzB1), len(nzB2)
```

```python
nB = np.concatenate((nxB1, nxB2, nyB1, nyB2, nzB1, nzB2))
print('len(nB) before unique = ', len(nB))
nB = np.unique(nB)
print('len(nB) after unique = ', len(nB))
```

```python
nB.shape
```

```python
rB = resx[nB, :]
rB.shape
```

```python
import numpy as np
import pickle
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_from_cpp.csv', header=None)
df.columns = ['x', 'y', 'z', 'h_before', 'h_after']
```

```python
print(df.shape)
df.head()
```

```python
x = df['x'].values
y = df['y'].values
z = df['z'].values

h_before = df['h_before'].values
h_after = df['h_after'].values
```

```python
plt.scatter(y, z, s = 0.001, color = 'k')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
```

```python
j = 500

dist = ((x - x[j])**2 + (y - y[j])**2 + (z - z[j])**2)**0.5

h_bef = 2.0 * h_before[j]
h_aft = 2.0 * h_after[j]

n_bef = np.sum(dist <= h_bef)
n_aft = np.sum(dist <= h_aft)

print(f'Before: h = {np.round(h_bef, 4)}, Nngb = {n_bef}')
print(f'After: h = {np.round(h_aft, 4)}, Nngb = {n_aft}')
```
